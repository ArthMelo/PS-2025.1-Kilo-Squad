```python
# Função para criação de features
def criar_features(df):
    df = df.copy()

    # Adicionar feature de jogo em casa
    # Time 1 é sempre em casa, e Time 2 fora
    df['Time casa 1'] = 1
    df['Time casa 2'] = 0

    # Diferença de posse de bola
    if 'Posse 1(%)' in df.columns and 'Posse 2(%)' in df.columns:
        df['Diferença de posse'] = df['Posse 1(%)'] - df['Posse 2(%)']

    # Diferença de chutes a gol
    if 'Chutes a gol 1' in df.columns and 'Chutes a gol 2' in df.columns:
        df['Diferença de chute a gol'] = df['Chutes a gol 1'] - df['Chutes a gol 2']

    # Diferença de finalizações totais (a gol + fora)
    if all(col in df.columns for col in ['Chutes a gol 1', 'Chutes a gol 2', 'Chutes fora 1', 'Chutes fora 2']):
        df['Finalizações 1'] = df['Chutes a gol 1'] + df['Chutes fora 1']
        df['Finalizações 2'] = df['Chutes a gol 2'] + df['Chutes fora 2']
        # Calcular a diferença
        df['Diferença de chutes'] = df['Finalizações 1'] - df['Finalizações 2']

    # Diferença de cartões
    if all(col in df.columns for col in ['Cartões amarelos 1', 'Cartões amarelos 2', 'Cartões vermelhos 1', 'Cartões vermelhos 2']):
        # Criar índice ponderado (vermelho vale mais que amarelo)
        df['Cartões 1'] = df['Cartões amarelos 1'] + 3 * df['Cartões vermelhos 1']
        df['Cartões 2'] = df['Cartões amarelos 2'] + 3 * df['Cartões vermelhos 2']
        df['Diferença cartões'] = df['Cartões 1'] - df['Cartões 2']

     # Índice de pressão ofensiva
    if all(col in df.columns for col in ['Escanteios 1', 'Escanteios 2']):
        df['Pressão ofensiva 1'] = df['Escanteios 1']
        df['Pressão ofensiva 2'] = df['Escanteios 2']
        
    # Adicionar chutes a gol
    if 'Chutes a gol 1' in df.columns and 'Chutes a gol 2' in df.columns:
        df['Pressão ofensiva 1'] += df['Chutes a gol 1']
        df['Pressão ofensiva 2'] += df['Chutes a gol 2']

    df['Diferença de pressão ofensiva'] = df['Pressão ofensiva 1'] - df['Pressão ofensiva 2']

    # Eficiência de chutes
    if all(col in df.columns for col in ['Chutes a gol 1', 'Chutes a gol 2', 'Finalizações 1', 'Finalizações 1']):
        df['Efiência de chutes 1'] = df['Chutes a gol 1'] / df['Finalizações 1'].replace(0, 1)
        df['Eficiência de chutes 2'] = df['Chutes a gol 2'] / df['Finalizações 2'].replace(0, 1)
        df['Diferença de eficiência de chutes'] = df['Efiência de chutes 1'] - df['Efiência de chutes 2']

    # Índice de faltas
    if all(col in df.columns for col in ['Faltas 1', 'Faltas 2']):
        df['Diferença de faltas'] = df['Faltas 1'] - df['Faltas 2']

    # Índice de eficiência ofensiva (chutes a gol / posse de bola)
    if all(col in df.columns for col in ['Chutes a gol 1', 'Chutes a gol 2', 'Posse 1(%)', 'Posse 2(%)']):
        df['Eficiência ofensiva 1'] = df['Chutes a gol 1'] / df['Posse 1(%)'].replace(0, 1)
        df['Eficiência ofensiva 2'] = df['Chutes a gol 2'] / df['Posse 2(%)'].replace(0, 1)
        df['Diferença de eficiência ofensiva'] = df['Eficiência ofensiva 1'] - df['Eficiência ofensiva 2']

    return df
```

```python
# Dataframe com as features criadas
df = criar_features(stats_sem_nulos)
```

```python
# Treinamento de Machine Learning

# Feito o tratamento de dados, nós podemos começar a fazer o treinamento do machine learning
## A gente vai precisar dividir os dados para treinamento e para teste

# Aqui há um problema que os modelos de regressão que usaremos não permitem strings, então irei tirar os nomes e a posição
# caso a acuracia não seja ideal, podemos definir como podemos adicionar elas para melhorar o resultado

X = stats_sem_nulos.drop(columns=['Resultado', 'Gols 1', 'Gols 2', 'Time 1', 'Time 2', 'Position 1', 'Position 2']) 
Y = stats_sem_nulos.Resultado # aqui ta a colunas que ele vai comparar pra ver se acertou


X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size=0.20, random_state = 42)  #separando 20% para o teste

```

```python
# Aqui está acontecendo a normalização dos dados
scaler = MinMaxScaler() 
X_treino_norma = scaler.fit_transform(X_treino)
X_teste_norma = scaler.transform(X_teste)
```

```python
# Aqui está o primeiro treinamento de Machine Learning, utilizando REGRESSÃO LINEAR
LR = LogisticRegression(random_state=0)
LR.fit(X_treino_norma, Y_treino)

LR.score(X_teste_norma, Y_teste)
```

```python
# Aqui está sendo usado o método de Árvore de Regressão para o machine learning

DT = DecisionTreeClassifier(random_state=42)
DT.fit(X_treino_norma, Y_treino) 

DT.score(X_teste_norma, Y_teste)
```

```python
# Aqui está sendo usado o método de Classificação de Árvore Aleatória para o machine learning

RF = RandomForestClassifier(n_estimators=100, min_samples_split=10, random_state=42)
RF.fit(X_treino_norma, Y_treino)

RF.score(X_teste_norma, Y_teste)
```

```python
# Vendo a importancia que o algoritmo deu para cada coluna

for i in zip(X.columns, DT.feature_importances_):
    print(i)
```

```python
for i in zip(X.columns, RF.feature_importances_):
    print(i)
```
